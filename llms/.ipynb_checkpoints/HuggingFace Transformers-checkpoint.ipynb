```python
from transformers import pipeline
```

```python
sentiment_classifier = pipeline("sentiment-analysis")
```

```python
sentiment_classifier("I am so excited to be learning about large language model")
```

```python
ner = pipeline("ner", model="dslim/bert-base-NER")
```

```python
ner("Her name is Anna and she works in New York City for Morgan Stanley")
```

```python
zeroshot_classifier = pipeline("zero-shot-classification", model = "facebook/bart-large-mnli")
```

```python
sequence_to_classify = "one day I will see the world"
candidate_labels = ['travel', 'cooking', 'dancing']
```

```python
zeroshot_classifier(sequence_to_classify, candidate_labels)
```

## Pre-trained Tokenizers

```python
from transformers import AutoTokenizer
```

```python
model = "bert-base-uncased"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
sentence = "I am so excited to learn large language model"
```

```python
input_ids = tokenizer(sentence)
print(input_ids)
```

```python
tokens = tokenizer.tokenize(sentence)
```

```python
print(tokens)
```

```python
tokenizer.decode(101)
```

```python
tokenizer.decode(102)
```

```python
model2 = "xlnet-base-cased"
```

```python
tokenizer2 = AutoTokenizer.from_pretrained(model2)
```

```python
input_ids = tokenizer2(sentence)
```

```python
print(input_ids)
```

```python
tokens = tokenizer2.tokenize(sentence)
print(tokens)
```

```python
tokenizer2.decode(4)
```

```python
tokenizer2.decode(3)
```

 ## Huggingface and Pytorch/Tensorflow

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
from transformers import logging
logging.set_verbosity_error()
```

```python
print(sentence)
print(input_ids)
```

```python
input_ids_pt = tokenizer(sentence, return_tensors = "pt")
print(input_ids_pt)
```

```python

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
input_ids_pt = tokenizer(sentence, return_tensors ="pt")
print(input_ids_pt)
```

```python
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
```

```python
with torch.no_grad():
    logits = model(**input_ids_pt).logits

predicted_class_id = logits.argmax().item()
model.config.id2label[predicted_class_id]
```
