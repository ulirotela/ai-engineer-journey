```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
data = pd.read_csv("tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data['Review'][0]
```

```python
data['review_lowercase'] = data['Review'].str.lower()
```

```python
data.head()
```

```python
en_stopwords = stopwords.words('English')
```

```python
en_stopwords.remove("not")
```

```python
data['review_no_stopwords'] = data['review_lowercase'].apply(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]))
```

```python
data.head()
```

```python
data['review_no_stopwords'][0]
```

```python
#PUNCTUATION
```

```python
#We want to remove punctation but we notice we have a few reviews with the symbol * instead of the word star.
```

```python
data['review_no_stopwords_no_star'] = data.apply(lambda x: re.sub(r"[*]","star", x['review_no_stopwords']), axis =1)
```

```python
data.head()
```

```python
data['review_no_stopwords_no_puntu'] = data.apply(lambda x:re.sub(r"([^\w\s])", "",x['review_no_stopwords_no_star']), axis=1)
```

```python
data.head()
```

```python
#TOKENIZING
```

```python
data['tokenized'] = data.apply(lambda x: word_tokenize(x['review_no_stopwords_no_puntu']), axis=1)
```

```python
data.head()
```

```python
data['tokenized'][0]
```

```python
ps = PorterStemmer()
```

```python
data["stemmed"] = data["tokenized"].apply(lambda tokens: [ps.stem(token) for token in tokens])
```

```python
data.head()
```

```python
data['stemmed'][0]
```

```python
data["lemmatized"] = data["tokenized"].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
data.head()
```

```python
tokens_clean = sum(data['lemmatized'], [])
```

```python
# unigrams: n=1
unigrams = (pd.Series \
            (nltk.ngrams(tokens_clean, 1)) \
            .value_counts()) 
print(unigrams)
```

```python
# bigrams: n=2
bigrams = (pd.Series \
           (nltk.ngrams(tokens_clean, 2)) \
           .value_counts()) 
print(bigrams)
```

```python
ngrams_4 = (pd.Series \
            (nltk.ngrams(tokens_clean, 4)) \
            .value_counts()) 
print(ngrams_4)
```
