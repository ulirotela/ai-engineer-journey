```python
# import packages
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```python
eng_stopwords = stopwords.words('english')
```

```python
print(eng_stopwords)
```

```python
sentence = "Orlando was too far to to the shop and he did not want her to walk alone"
```

```python
#keep the words in the sentance if the word is not in the list of stop words
sentence_no_stopwords = ' '.join([word for word in sentence.split() if word not in (eng_stopwords)])
print(sentence_no_stopwords)
```

```python
#removing stopwords from the list
eng_stopwords.remove("did")
eng_stopwords.remove("not")
```

```python
#removing stop words from the list
eng_stopwords.append("go")
```

```python
sentence_no_stopwords_custom = ' '.join(word for word in sentence.split() if word not in (eng_stopwords))
print(sentence_no_stopwords_custom)
```
