```python
#The next step in preprocessing is to standardise the text. One option for this is stemming, where words are reduced to their base form. For example, words like ‘connecting’ or ‘connected’ 
#will be stemmed to the base form ‘connect’
```

```python
from nltk.stem import PorterStemmer
```

```python
# create stemmer
ps = PorterStemmer()
```

```python
connect_tokens = ['connecting', 'connected', 'connectivity', 'connect', 'connects']
```

```python
for t in connect_tokens:
    print(t, " : ", ps.stem(t))
```

```python
learn_tokens = ['learned', 'learning', 'learn', 'learns', 'learner', 'learners']
```

```python
for t in learn_tokens:
    print(t, " : ", ps.stem(t))
```
